#Implementing TF-IDF in Python


```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

##Term frequency from scratch

```python
import numpy as np
import pandas as pd
text = ["The quick brown fox","The dog.","The fox"]

#sentences and tokenization
word_data = [sentences.split() for sentences in text]

#finding vocab by finding unique words in the sentences of word_data
vocab = set([word for sentences in word_data for word in sentences])

tf_data = []
#term frequency
for sentences in word_data:
  tf_sentences = []
  for word in vocab:
    tf_sentences.append(sentences.count(word)/len(sentences))
  tf_data.append(tf_sentences)

print(tf_data)

#finding inverse document frequency
n_docu = len(text)
idf_data = []

# Ensure vocab is ordered for consistent IDF calculation with TF data
sorted_vocab = sorted(list(vocab))

for word in sorted_vocab:
  n_apperances = 0;
  for sentences in word_data:
    if word in sentences:
      n_apperances += 1

  #print(word,n_apperances)

  idf = np.log(n_docu/n_apperances)
  idf_data.append(idf)

print(idf_data)

#finding TF - IDF
tfidf_data = []
for i, doc_tf_list in enumerate(tf_data):
    doc_tfidf = []
    for j, tf_value in enumerate(doc_tf_list):
        doc_tfidf.append(tf_value * idf_data[j])
    tfidf_data.append(doc_tfidf)

print(tfidf_data)
df = pd.DataFrame(tfidf_data, columns=sorted_vocab)
df
```

## TF-IDF with Sklearn




```python
text = ["The quick brown fox","The dog.","The fox"]

vec = TfidfVectorizer()
vec.fit(text)
tfidf_array = vec.transform(text).toarray()

# Get feature names for column headers
feature_names = vec.get_feature_names_out()

# Create a DataFrame
tfidf_df = pd.DataFrame(tfidf_array, columns=feature_names)
display(tfidf_df)
```
